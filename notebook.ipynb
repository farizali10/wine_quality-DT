```python
dt_model1 = DecisionTreeClassifier()
```

```python
dt_model1.fit(x1,y1)
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(dt_model1,filled=True)
```

## Inferences:
- Samples in the box1 mean no of datapoints we have considered.
- Values in box1 mean that we had 4 classes i.e. 4,5,6,7 in quality column (output column) so values = [1,21,5,3] means we have:
    - 1 value of class 4
    - 21 values of class 5
    - 5 values of class 6
    - 3 values of class 7
     
    
    <br>Total of all values will result in total sample values i.e. 1 + 21 + 5 + 3 = 30 

- x[1] here means the column which is the parent node i.e. 2nd column of x dataframe which shows it has the lowest gini impurity

    - Based on x[2] = 0.335 (threshold) of x[2] column values if greater than it will got right side else left
- 26 samples in right side split shows 26 values greater than x[2]'s 0.335.
- Now in left hand node the same thing is repeated started from again calculating gini.


```python
# Plot with column and class names

plt.figure(figsize=(20,20))
tree.plot_tree(dt_model1,filled=True,class_names = [str(i)for i in set(y1)], feature_names = x1.columns)
```

- Performing Post Prunning on level 3 where there are 20 records for class 5 and 1 record for class 6.  
    - Removing level 3 to control complexity of the tree


```python
# Checking train and test accuracy
dt_model1.predict(x1)
```

```python
dt_model1.score(x1,y1)
```

```python
dt_model1.score(x_test,y_test)
```

Here the model is overfitted as we are getting higher accuracy and on random sample as input its giving lower accuracy.

- Here lower accuracy isnt a great issue as compared to overfitting so we would resolve overfitting here.

# FineTuning CPP:
by default cpp = 0 as it would create the most separations in a tree i can so we would try to control complexity of the model by choosing best CPP.

```python
path = dt_model1.cost_complexity_pruning_path(x1,y1)
```

```python
path
```

```python
ccp_alpha = path.ccp_alphas
```

```python
# Passsing different ccp alphas
dt_model2 = []
for ccp in ccp_alpha:
    dt_m = DecisionTreeClassifier(ccp_alpha=ccp)
    dt_m.fit(x1,y1)
    dt_model2.append(dt_m)
```

```python
dt_model2
```

First one has ccp_alpha = 0
<br>

In total we have built 8 models so far

```python
"""train_score = dt_model1.score(x1,y1) fo
test_score = dt_model1.score()"""
```

```python
train_scoreCPP = [i.score(x1,y1) for i in dt_model2]
```

```python
test_scoreCPP = [i.score(x_test,y_test) for i in dt_model2]
```

```python
test_scoreCPP
```

```python
train_scoreCPP
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alpha, train_scoreCPP, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alpha, test_scoreCPP, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```

Here the accuracy is low but we are still doing some progress in terms of reducing overfitting but the overfitting is due to the fact that our training dataset is very small and testing dataset is large so our model wasnt even able to learn patterns of test data that are unavailable completely in training data.

Grid and Random Search CV will try to find best parameters value in sklearn.decisiontreeclassifier just like how we had tuned "ccp" it will tune other parameters to find best parameter.

But not one by one just like how we did as it would take time so it will Grid and Random Search CV will automatically select the best hyperparameter based upon cross validation (CV) split that it will do by dividing "TRAINING" data into 5 different parameters where it will divide 4 dataset into training and 1 into testing and it will keep on doing that.

```python
grid_param = {
    "criterion" : ["gini","entropy"],
    "splitter" : ["best","random"],
    "max_depth" : range(2,40,1),
    "min_samples_split" : range(2,10,1),
    "min_samples_leaf" : range(1,10,1),
}
```

```python
grid_ccp = GridSearchCV(estimator=dt_model_ccp, param_grid=grid_param, cv=5, n_jobs=-1)
```

```python
grid_ccp.fit(x1,y1)
```

```python
grid_ccp.best_params_
```
